```python
# 2.basic data exploratory analysis:
    # 2.1.missing value treatment
col_to_drop = ['Intersection Street 1','Intersection Street 2','Landmark','School or Citywide Complaint','Vehicle Type','Taxi Company Borough',
               'Taxi Pick Up Location','Bridge Highway Name','Bridge Highway Direction','Road Ramp','Bridge Highway Segment','Garage Lot Name',
               'Ferry Direction','Ferry Terminal Name']
df.drop(columns = col_to_drop, inplace = True)
```

```python
    # 2.2.1. Draw a frequency plot for city-wise complaints
city_counts = df['City'].value_counts()
print(city_counts)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
city_counts.plot(kind='bar')
plt.title('Complaints across cities')
plt.xlabel('city')
plt.ylabel('count')
```

```python
    # 2.2.2.Draw scatter and hexbin plots for complaint concentration across Brooklyn
brooklyn_complaint = df[df['City'] == 'Brooklyn']
print(brooklyn_complaint)
```

```python
# Create a scatter plot for complaint concentration in Brooklyn
plt.figure(figsize=(10, 8))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, marker='o', c='b')
plt.title('Complaint Concentration Across Brooklyn')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()
```

```python
# Create a hexbin plot for complaint concentration in Brooklyn
plt.figure(figsize=(12, 8))
plt.hexbin(df['Longitude'], df['Latitude'], gridsize=30, cmap='Blues')
plt.title('Hexbin Plot of Complaint Concentration Across Brooklyn')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.colorbar()
plt.show()
```

```python
# 3.3.Display the types of complaints in each city in a separate dataset
group = df.groupby('City')
```

```python
city_complaint_data = {}
```

```python
#5. Check if the average response time across various types of complaints
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['Closed Date'] = pd.to_datetime(df['Closed Date'])

# Calculate response time (time taken to close the complaint)
df['Response Time'] = df['Closed Date'] - df['Created Date']

# Calculate the average response time for each complaint type
average_response_time = df.groupby('Complaint Type')['Response Time'].mean()

print(average_response_time)
```
